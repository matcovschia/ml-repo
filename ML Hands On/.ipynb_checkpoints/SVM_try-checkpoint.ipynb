I want to create three labels, which I would later use label encoding on, to represent the portion of bookings per each country.
I was thinking of doing this in the following manner : 
        Country Percentage > 10%  High 
 10% >= Country Percentage > 1%   Medium
 1%  >= Country Percentage        Low
 
This idea came from user Shrikant3542, found at https://www.kaggle.com/shreekant009. Though I do end up using a different scale and portion of the data set. Ultimately our methods arent exactly the same, at least in regards to the encoding of a couple of the categorical data columns.

```python
x=((bookings['country'].value_counts()/len(bookings['country']))*100).to_dict()
x
```

```python
for i in x:
    percentage = x.get(i,0)
    key= str(i)
    if percentage >=10:
        bookings.country.replace(i,'High',inplace=True)
    elif percentage >=1:
        bookings.country.replace(i,'Medium',inplace=True)
    else:
        bookings.country.replace(i,'Low',inplace=True)

a = bookings.country.isna().sum()
b = len(bookings)
percentage = a/b
if percentage >=10:
        bookings.country.loc[bookings.country.isnull()] = 'High'
elif percentage >=1:
        bookings.country.loc[bookings.country.isnull()] = 'Medium'
else:
        bookings.country.loc[bookings.country.isnull()] = 'Low'
```

It has come to my attention that a couple of the couples aren't factors, and disrupt the working of most models, these columns are;is_canceled(for obvious reasons), reservation_status,previous_cancellations,previous_bookingS_not_canceled. Is_canceled is what we're trying to predict, including it would be foolish. Reservation_status is more or less the same as Is_canceled, Check-Out means no cancel, Canceled means canceled and No-Show means neither. Both of the previous columns don't really mean anything, as in this case I am considering each booking an independent bernoulli trial, just because a person has canceled the previous 4 bookings doesnt increase the chance of that person cancelling again or not cancelling again.

```python
y=bookings.is_canceled
x=bookings.drop(["is_canceled"],axis=1)
```

```python
bookings_1 = bookings[bookings["is_canceled"]==1]
bookings_0 = bookings[bookings["is_canceled"]==0]
print("Total Rows : ", len(bookings))
print("% of zeroes : ", len(bookings_0)/len(bookings)*100)
print("% of ones : ", len(bookings_1)/len(bookings)*100)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=.2,random_state=11)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
def display_scores(scores):
    print("Scores            :", scores)
    print("Mean              :", scores.mean())
    print("Standard Deviation:", scores.std())
    
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(x,y)
booking_prediction_for = forest_reg.predict(x)
forest_mse = mean_squared_error(y,booking_prediction_for)
forest_rmse = np.sqrt(forest_mse)
```

```python
forest_rmse
```

```python
forest_scores = cross_val_score(forest_reg,x,y,scoring="neg_mean_squared_error",cv=10 )
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
booking_prediction_lin = lin_reg.predict(x)
lin_mse = mean_squared_error(y,booking_prediction_lin)
lin_rmse = np.sqrt(lin_mse)
```

```python
lin_rmse
```

```python
lin_scores = cross_val_score(lin_reg,x,y,scoring="neg_mean_squared_error",cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(x,y)
booking_prediction_tree = tree_reg.predict(x)
tree_mse = mean_squared_error(y,booking_prediction_tree)
tree_rmse = np.sqrt(tree_mse)
```

```python
tree_rmse
```

```python
scores = cross_val_score(tree_reg,x,y, scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)

display_scores(tree_rmse_scores)
    
```
